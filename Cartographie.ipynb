```python
from IPython.display import display, IFrame, Javascript
from ipywidgets import Box,Dropdown,HTML,VBox,Image,Label,Text,HBox,Layout,GridspecLayout,FileUpload,Button
# Permet de tout executer au lancement du notebook + conserver le notebook actif pendant 2h
from masquer import *

display(Javascript("""
function repeter(){
IPython.notebook.kernel.execute("a=1");
}
// execute a = 1 en python toutes les 8 minutes pendant 2h
let timerId = setInterval(() => repeter(), 480000);
setTimeout(() => { clearInterval(timerId); alert('fin de cession'); }, 7200000);

// Supprimer la taille limite pour la sortie d'une cellule
IPython.OutputArea.prototype._should_scroll = function(lines) {
    return false;
};
IPython.notebook.kernel.execute("url = '" + window.location + "'");"""))

f = FileUpload(
    description = "CHARGER",
    accept='.snt',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False  # True to accept multiple files upload else False
)

bou = Button(
    description='SAUVER',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='SAUVER',
    icon='' # (FontAwesome names without the `fa-` prefix)
)

save = {}
save_state = {}
saved_widgets = {}
charge = False

def sauver_etats(b=None):
    global save_state,save,saved_widgets,charge
    if not charge:
        k = saved_widgets.keys()
        for val in k:
            save_state[val]=saved_widgets[val].value
        save['Cartographie']=save_state
        with open('SAVE.snt', 'w') as out:
            out.write(str(save))
    
def charger_etats(b=None):
    global save_state,saved_widgets,save,charge
    charge=True
    with open('SAVE.snt', 'a') as file:
        pass
    with open('SAVE.snt', 'r') as file:
        save2 = {}
        try:
            save2=eval(file.read())
        except:
            pass
    save = save2
    save_state2 = {}
    try:
        save_state2=save2['Cartographie']
    except:
        pass
    k = save_state2.keys()
    for val in k:
        if save_state2[val] != '':
            save_state[val] = save_state2[val]
    k = save_state.keys()
    for val in k:
        try:
            saved_widgets[val].value = save_state[val]
        except:
            pass
    save['Cartographie'] = save_state
    charge = False 
    sauver_etats()
    
    
def on_click(b):
    url = "../notebooks/SAVE.snt"
    display(Javascript('window.open("{url}");'.format(url=url)))
         
def on_use(b):
    global save_state
    if f.data ==[]:
        return
    
    with open("SAVE.snt", "w+b") as i:
        i.write(f.data[0])
    charger_etats()
    f.data = []

display(HTML("<h1 style='text-align:center;'>PERSONNALISATION DE VOTRE LIVRE NUMERIQUE</h1>"))
display(HBox([bou,f],layout=Layout(width='100%',justify_content="center")))

    
bou.on_click(on_click)  
f.observe(on_use,"data")
```

Les cartes ont été systématiquement numérisées à la fin du XXe siècle.  


Le principal instrument de localisation, GPS (Global Positioning System), a été conçu par l’armée américaine. Le premier satellite GPS fut lancé **en 1978**. Il y en a actuellement **une trentaine**, de sorte qu’à tout moment **quatre à six satellites au moins** sont visibles depuis tout point de la Terre. Couplé aux cartes numériques le système **GPS** permet de se situer. Il n’est pas toujours efficace en ville et peut être complété par d’autres moyens de localisation comme la détection de bornes Wi-Fi proches. 


D’autres systèmes plus précis, dont **Galileo**, sont en cours de déploiement.
